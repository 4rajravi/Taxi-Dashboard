```python
import pandas as pd
import numpy as np
import zipfile

import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import cm

import seaborn as sns
sns.set(style="white", palette="muted", color_codes=True)

#from haversine import haversine
from scipy.spatial.distance import euclidean , cityblock
from geopy.distance import great_circle
from math import *

from bokeh.io import output_notebook,show
from bokeh.models import HoverTool
from bokeh.plotting import figure
from bokeh.palettes import Spectral4

import folium 
from folium import plugins
from folium.plugins import HeatMap

import warnings
warnings.filterwarnings("ignore")
```

```python
zip_path = 'data/T-drive Taxi Trajectories.zip'  # Path to the ZIP file
target_file = 'release/taxi_log_2008_by_id/10.txt'         # Name of the file inside the ZIP

with zipfile.ZipFile(zip_path, 'r') as z:
    with z.open(target_file) as f:
        data = pd.read_csv(f, header=None, names=['taxi_id', 'timestamp', 'longitude', 'latitude'])

data.head()
```

```python
data.time = pd.to_datetime(data.timestamp)
```

```python
data["Coordinates"] = list(zip(data.latitude , data.longitude))
data.head()
```

```python
data.isna().mean()
```

```python
print("Longitude range:",data.longitude.max(),data.longitude.min())
```

```python
print("Latitude range:",data.latitude.max(),data.latitude.min())
```

```python
import folium
mapit = None
latlon = list(data.Coordinates)
for coord in latlon:
    mapit = folium.Map( location=[ coord[0], coord[1] ] )

mapit.save( 'map1.html')
```
